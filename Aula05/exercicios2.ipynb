```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lojas e Estoque
lojas = ['A', 'B', 'C', 'D']
estoque = [500, 100, 300, 600]

# Vendas em 2022 por mÃªs
vendas_2022 = {
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
# a

df_estoque = pd.DataFrame({
    'Loja': lojas,
    'Estoque': estoque
}, index=['AA', 'BB', 'CC', 'DD'])

print(df_estoque)
```

```python
# b

df_vendas = pd.DataFrame(vendas_2022, index=['AA', 'BB', 'CC', 'DD'])
df_vendas['Loja'] = lojas
df_vendas = df_vendas.set_index('Loja')

print(df_vendas)
```

```python
# c

fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.suptitle('Janeiro - Vendas e Estoque por Loja')

janeiro = vendas_2022['Jan']

for i, ax in enumerate(axs.flatten()):
    ax.bar(['Vendas', 'Estoque'], [janeiro[i], estoque[i]], color=['#ADD8E6', 'pink'])
    ax.set_title(f'Loja {lojas[i]}')

plt.tight_layout()
plt.show()
```

```python
# Cores para as lojas A, B, C, D
cores_lojas = {
    'A': '#ADD8E6',  # azul
    'B': '#5d6ab2',  # lavanda
    'C': '#8ddb8a',  # verde
    'D': '#d76d64'   # vermelho
}
```

```python
# d

# Plotando grÃ¡fico de barras verticais (colunas)
df_vendas.T.plot(kind='bar',
                 figsize=(12, 6),
                 color=[cores_lojas[loja] for loja in df_vendas.index],
                 title='Vendas Mensais por Loja')

plt.xlabel('MÃªs')
plt.ylabel('Quantidade Vendida')
plt.legend(title='Lojas')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# e

df_vendas.T.plot(kind='line',
                 figsize=(12, 6),
                 marker='o',
                 color=[cores_lojas[loja] for loja in df_vendas.index],
                 title='EvoluÃ§Ã£o de Vendas por Loja')

plt.xlabel('MÃªs')
plt.ylabel('Quantidade Vendida')
plt.legend(title='Lojas')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# f

vendas_totais = df_vendas.sum(axis=1)

vendas_totais.plot(kind='barh', color='#650c68', title='Total de Vendas por Loja')
plt.xlabel('Quantidade Total Vendida')
plt.ylabel('Loja')
plt.grid(True)
plt.tight_layout()
plt.show()
```
