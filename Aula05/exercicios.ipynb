```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
# Realizar a leitura do meu arquivo csv
df = pd.read_csv('base de dados/monitoramento_tempo.csv',sep=',')
```

```python
df["data"] = pd.to_datetime(df["data"])
```

```python
# a
plt.figure(figsize=(15,8))
plt.plot(df['data'], df['humidade'])
plt.xlabel('Ano', fontsize =15)
plt.ylabel('Umidade', fontsize =15)
plt.title('Umidade no momento', fontsize=20, pad=20)
```

```python
# b
plt.figure(figsize=(15,8))
plt.plot(df['data'], df['velocidade do vento'])
plt.xlabel('Ano', fontsize =15)
plt.ylabel('Velocidade do vento', fontsize =15)
plt.title('Velocidade do vento no momento', fontsize=20, pad=20)
```

```python
#c 

# realizando anotaes no grfico
fig = plt.figure(figsize=(15,5))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['humidade'], color='b')
eixo.grid(True)
eixo.set_title('Umidade no momento', fontsize=25, pad=20)
eixo.set_ylabel('Umidade', fontsize=20)
eixo.set_xlabel('Ano', fontsize=20)
eixo.legend(['Umidade'], loc='best')

# Linhas tracejadas do mximo e mnimo
eixo.axhline(max(df['humidade']), color='k', linestyle='--')
eixo.axhline(min(df['humidade']), color='k', linestyle='--')

# Anotao do valor MXIMO
x_max = df['data'][df['humidade'].idxmax()]
y_max = max(df['humidade'])

eixo.annotate('Mximo',
              xy=(x_max, y_max),
              xytext=(x_max, y_max + 10),  # posio do texto acima
              textcoords='data',
              fontsize=15,
              ha='center',
              arrowprops=dict(facecolor='k', arrowstyle='->'))

# Anotao do valor MNIMO
x_min = df['data'][df['humidade'].idxmin()]
y_min = min(df['humidade'])

eixo.annotate('Mnimo',
              xy=(x_min, y_min),
              xytext=(x_min, y_min - 10),  # posio do texto abaixo
              textcoords='data',
              fontsize=15,
              ha='center',
              arrowprops=dict(facecolor='k', arrowstyle='->'))
```

```python
#d

# realizando anotaes no grfico
fig = plt.figure(figsize=(15,5))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['direo do vento'], color='purple')
eixo.grid(True)
eixo.set_title('direo do vento no momento', fontsize=25, pad=50)
eixo.set_ylabel('direo do vento', fontsize=20)
eixo.set_xlabel('Ano', fontsize=20)
eixo.legend(['direo do vento'], loc='upper right')

# Linhas tracejadas do mximo e mnimo
eixo.axhline(max(df['direo do vento']), color='k', linestyle='--')
eixo.axhline(min(df['direo do vento']), color='k', linestyle='--')

# Anotao do valor MXIMO
x_max = df['data'][df['direo do vento'].idxmax()]
y_max = max(df['direo do vento'])

eixo.annotate('Mximo',
              xy=(x_max, y_max),
              xytext=(x_max, y_max + 30), 
              textcoords='data',
              fontsize=15,
              ha='center',
              arrowprops=dict(facecolor='k', arrowstyle='->'))

# Anotao do valor MNIMO
x_min = df['data'][df['direo do vento'].idxmin()]
y_min = min(df['direo do vento'])

eixo.annotate('Mnimo',
              xy=(x_min, y_min),
              xytext=(x_min, y_min - 30),
              textcoords='data',
              fontsize=15,
              ha='center',
              arrowprops=dict(facecolor='k', arrowstyle='->'))
```
