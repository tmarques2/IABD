```python
#Exercicio 3

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dados fornecidos
A = np.array([50, 100, 120, 300]).reshape(-1, 1)
Valor = np.array([180000, 300000, 375000, 600000])

# Criar e treinar o modelo
modelo = LinearRegression()
modelo.fit(A, Valor)

# Teste: prever valores para áreas
A_teste = np.array([80, 150, 250]).reshape(-1, 1)
Valor_previsto = modelo.predict(A_teste)
```

```python
# Resultados
for i in range(len(A_teste)):
    print(f"Área: {A_teste[i][0]} m² - Valor previsto: R$ {Valor_previsto[i]:.2f}")

# Plotar
plt.scatter(A, Valor, color='blue', label='Dados reais')
plt.plot(A_teste, Valor_previsto, color='red', label='Previsão')
plt.xlabel('Área (m²)')
plt.ylabel('Valor do Imóvel (R$)')
plt.title('Previsão de Valor de Imóvel')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Exercicio 4

# Dados fornecidos
Renda = np.array([1200, 5000, 8000, 10000, 30000]).reshape(-1, 1)
Gastos = np.array([750, 1500, 4000, 6000, 18000])

# Criar e treinar o modelo
modelo = LinearRegression()
modelo.fit(Renda, Gastos)

# Teste: prever valores para novas rendas
Renda_teste = np.array([2000, 7000, 15000, 25000]).reshape(-1, 1)
Gastos_previstos = modelo.predict(Renda_teste)
```

```python
# Resultados
for i in range(len(Renda_teste)):
    print(f"Renda: R$ {Renda_teste[i][0]} - Gasto previsto: R$ {Gastos_previstos[i]:.2f}")

# Plotar
plt.scatter(Renda, Gastos, color='green', label='Dados reais')
plt.plot(Renda_teste, Gastos_previstos, color='red', label='Previsão')
plt.xlabel('Renda (R$)')
plt.ylabel('Gastos (R$)')
plt.title('Previsão de Gastos com base na Renda')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Exercicio 5

# Dados fornecidos
Consumo = np.array([50, 140, 30, 75, 123]).reshape(-1, 1)
Demanda = np.array([100, 200, 80, 100, 160])

# Criar e treinar o modelo
modelo = LinearRegression()
modelo.fit(Consumo, Demanda)

# Teste: prever valores para novos consumos
Consumo_teste = np.array([60, 90, 150]).reshape(-1, 1)
Demanda_prevista = modelo.predict(Consumo_teste)
```

```python
# Plotar
plt.scatter(Consumo, Demanda, color='purple', label='Dados reais')
plt.plot(Consumo_teste, Demanda_prevista, color='red', label='Previsão')
plt.xlabel('Consumo (KW)')
plt.ylabel('Demanda')
plt.title('Previsão de Demanda de Energia')
plt.legend()
plt.grid(True)
plt.show()
```
