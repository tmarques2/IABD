```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados_x = np.linspace(0,10,10)
dados_y = dados_x * dados_x
```

```python
plt.figure(figsize=(10,8), facecolor='gray')
ax = plt.axes()
ax.set_facecolor('white')
plt.plot(dados_x, dados_y, color='red', marker='o', linestyle='dashed', markeredgecolor='black')
plt.xlabel('Eixo x', fontsize=16)
plt.ylabel("Eixo y", fontsize=16)
plt.title('Gráfico simples com Matplotlib', fontsize=16)
```

```python
#Grafico 1
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(12,12))
linhas = 2
colunas = 2
plt.subplot(linhas, colunas, 1)
plt.plot(dados_x, 'r--', label='dados_x')
plt.legend(loc='lower right')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Gráfico 1')

#Grafico 2
plt.subplot(linhas, colunas, 2)
plt.plot(dados_x, dados_y, 'b--', label='Dados x \n e y')
plt.title("Gráfico 2")
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()

#Grafico 3
plt.subplot(linhas, colunas, 3)
plt.plot(dados_x*dados_x, 'g-', label='dados_x^2')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()

#Grafico 4
plt.subplot(linhas,colunas,4)
plt.plot(dados_y,'c--',label='dados_y')
plt.title('Grafico 4')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()

plt.savefig('Graficos aula03.png') # serve para salvar figura
plt.show() # remove o warning
```

```python
#criando grafico de barras

figura = plt.figure()
grafico = figura.add_axes([0,0,0.7,0.7])
rotulos = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
# Plota o gráfico
grafico.bar(rotulos,dados_x,color='red') # plota grafico de barras
grafico.set_xlabel('Rotulos') # adicionando titulos nos eixos
grafico.set_ylabel('Quantidade')
```

```python
# Criando outro grafico de barras
grafico.bar(rotulos, dados_x,color='red') # Dados 1
grafico.bar(rotulos,dados_y*0.1,color='blue') # Dados 2
figura
```

```python
grafico.bar(dados_x, dados_y/2, color=['red', 'blue'])
figura
```
