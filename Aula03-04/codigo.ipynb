```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados_x = np.linspace(0,10,10)
dados_y = dados_x * dados_x
```

```python
plt.figure(figsize=(10,8), facecolor='gray')
ax = plt.axes()
ax.set_facecolor('white')
plt.plot(dados_x, dados_y, color='red', marker='o', linestyle='dashed', markeredgecolor='black')
plt.xlabel('Eixo x', fontsize=16)
plt.ylabel("Eixo y", fontsize=16)
plt.title('Gr√°fico simples com Matplotlib', fontsize=16)
```

```python
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(12,12))
linhas = 2
colunas = 2
plt.subplot(linhas, colunas, 1)
plt.plot(dados_x, 'r--', label='dados_x')
plt.legend(loc='lower right')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Gr√°fico 1')

plt.subplot(linhas, colunas, 2)
plt.plot(dados_x, dados_y, 'b--', label='Dados x \n e y')
plt.title("Gr√°fico 2")
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()
```
